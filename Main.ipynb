```python
%%writefile lookup-cpu.cpp

#include <array>
#include <cmath>
#include <cstdint>
#include <cstdio>
#include <iostream>
#include <memory>
#include <sstream>
#include <string>
#include <vector>
#include <stdlib.h>
#include <string.h>

uint64_t fnv1a64(const std::string &s) {
    uint64_t hash = 1469598103934665603ULL; 
    for (unsigned char c : s) {
        hash ^= c;
        hash *= 1099511628211ULL; 
    }
    return hash ? hash : 1ULL; 
}

uint64_t next_power_of_two(uint64_t x) {
    return 1ULL << (64 - __builtin_clzll(x - 1));
}

std::vector<uint64_t> build_dict_table(const std::vector<std::string> &mwes) {
    size_t dict_count = mwes.size();
    size_t table_size = next_power_of_two((size_t)(dict_count / 0.6)); 
    std::vector<uint64_t> dict_table(table_size, 0ULL);

    for (const std::string &mwe : mwes) {
        uint64_t f = fnv1a64(mwe);
        size_t mask = table_size - 1;
        size_t idx = f & mask;
        while (dict_table[idx] != 0ULL) {
            idx = (idx + 1) & mask; 
        }
        dict_table[idx] = f;
    }

    std::cout << "Built hash table with " << table_size << " slots.\n";
    return dict_table;
}

bool lookup_cpu(const std::vector<uint64_t> &table, uint64_t key) {
    size_t mask = table.size() - 1;
    size_t idx = key & mask;
    for (size_t probe = 0; probe < table.size(); ++probe) {
        uint64_t val = table[idx];
        if (val == 0ULL) return false;     
        if (val == key) return true;       
        idx = (idx + 1) & mask;            
    }
    return false; 
}

std::string join_words(const std::vector<std::string> &words, size_t start, size_t end) {
    std::ostringstream oss;
    for (size_t i = start; i <= end; ++i) {
        if (i > start) oss << " ";
        oss << words[i];
    }
    return oss.str();
}

std::vector<std::vector<bool>> create_lookup_table_cpu(const std::vector<std::string> &words,
                                    const std::vector<uint64_t> &dict_table) {
    size_t n = words.size();
    std::vector<std::vector<bool>> result(n, std::vector<bool>(n, false));

    for (size_t row = 0; row < n; ++row) {
        for (size_t col = row; col < n; ++col) { 
            std::string phrase = join_words(words, row, col);
            uint64_t key = fnv1a64(phrase);
            result[row][col] = lookup_cpu(dict_table, key);
        }
    }

    return result;
}

std::vector<std::string> split_sentence(const std::string &sentence) {
    std::vector<std::string> words;
    std::istringstream iss(sentence);
    std::string word;
    
    while (iss >> word) {
        words.push_back(word);
    }
    
    return words;
}

std::string clean_string(const std::string &input) {
    std::string result;
    result.reserve(input.size()); 

    for (char c : input) {
        if (!std::ispunct(static_cast<unsigned char>(c))) { 
            result += std::tolower(static_cast<unsigned char>(c)); 
        }
    }

    return result;
}

void read_sentences_from_file(std::vector<std::string> &sentences) {
    FILE* f = fopen("long_sentences.txt", "r");
    char buffer[1024];

    if (!f) return;

    while (fgets(buffer, sizeof(buffer), f)) {
        buffer[strlen(buffer) - 1] = '\0';
        sentences.push_back(std::string(buffer));
    }

    fclose(f);
}

void read_mwes_from_file(std::vector<std::string> &mwe) {
    FILE* f = fopen("mwes.txt", "r");
    char buffer[1024];

    if (!f) return;

    while (fgets(buffer, sizeof(buffer), f)) {
        buffer[strlen(buffer) - 2] = '\0';
        mwe.push_back(std::string(buffer));
    }

    fclose(f);
}

int main() {
    std::vector<std::string> sample_sentences;
    read_sentences_from_file(sample_sentences);
    std::vector<std::string> mwes; 
    read_mwes_from_file(mwes);

    auto dict_table = build_dict_table(mwes);
    double elapsed = 0;

    for (int i = 0; i < sample_sentences.size(); i++){
        auto sentence = sample_sentences[i];
        std::cout << "Processing - " << sentence << "\n";
        auto words = split_sentence(clean_string(sentence));
        clock_t start = clock();
        auto lookup_table = create_lookup_table_cpu(words, dict_table);
        clock_t end = clock();
        elapsed += (double)(end - start) / CLOCKS_PER_SEC;
        for (int x = 0; x < words.size(); x++){
            for (int y = 0; y < words.size(); y++){
                if (lookup_table[x][y] == 1){
                    std::cout << "Found MWE from index " << x << "-" << y << " - " << join_words(words, x, y) << "\n";
                }
            }
        }

        std::cout << "\n";
    }
    printf("Time: %f seconds\n", elapsed);

    return 0;
}
```

```python
%%writefile lookup-gpu.cu

#include <array>
#include <cmath>
#include <cstdint>
#include <cstdio>
#include <iostream>
#include <memory>
#include <sstream>
#include <string>
#include <vector>
#include <stdlib.h>
#include <string.h>

uint64_t fnv1a64(const std::string &s) {
    uint64_t hash = 1469598103934665603ULL; 
    for (unsigned char c : s) {
        hash ^= c;
        hash *= 1099511628211ULL; 
    }
    return hash ? hash : 1ULL; 
}

uint64_t next_power_of_two(uint64_t x) {
    return 1ULL << (64 - __builtin_clzll(x - 1));
}

std::vector<uint64_t> build_dict_table(const std::vector<std::string> &mwes) {
    size_t dict_count = mwes.size();
    size_t table_size = next_power_of_two((size_t)(dict_count / 0.6)); 
    std::vector<uint64_t> dict_table(table_size, 0ULL);

    for (const std::string &mwe : mwes) {
        uint64_t f = fnv1a64(mwe);
        size_t mask = table_size - 1;
        size_t idx = f & mask;
        while (dict_table[idx] != 0ULL) {
            idx = (idx + 1) & mask; 
        }
        dict_table[idx] = f;
    }

    std::cout << "Built hash table with " << table_size << " slots.\n";
    return dict_table;
}

__device__ bool lookup_gpu(const uint64_t *table, size_t table_size, uint64_t key) {
    size_t mask = table_size - 1;
    size_t idx = key & mask;
    for (int probe = 0; probe < 8; ++probe) { 
        uint64_t val = __ldg(&table[idx]);    
        if (val == 0ULL) return false;
        if (val == key) return true;
        idx = (idx + 1) & mask;
    }
    return false;
}

__global__ void lookup_gpu_multiple(const uint64_t *table, size_t table_size,
                                   const uint64_t *queries, bool *results, int n_queries) {
    int i = blockIdx.x * blockDim.x + threadIdx.x;
    int stride = blockDim.x * gridDim.x;
    
    for (; i < n_queries; i += stride)
        results[i] = lookup_gpu(table, table_size, queries[i]);
}

std::string join_words(const std::vector<std::string> &words, size_t start, size_t end) {
    std::ostringstream oss;
    for (size_t i = start; i <= end; ++i) {
        if (i > start) oss << " ";
        oss << words[i];
    }
    return oss.str();
}

std::vector<std::string> split_sentence(const std::string &sentence) {
    std::vector<std::string> words;
    std::istringstream iss(sentence);
    std::string word;
    
    while (iss >> word) {
        words.push_back(word);
    }
    
    return words;
}

std::vector<std::vector<bool>> create_lookup_table_gpu(
    const std::vector<std::string> &words,
    const std::vector<uint64_t> &dict_table)
{
    size_t n = words.size();
    std::vector<std::string> phrases;
    std::vector<uint64_t> keys;

    for (size_t row = 0; row < n; ++row) {
        for (size_t col = row; col < n; ++col) {
            std::string phrase = join_words(words, row, col);
            phrases.push_back(phrase);
            keys.push_back(fnv1a64(phrase));
        }
    }

    uint64_t *d_table, *d_queries;
    bool *d_results;
    size_t table_size = dict_table.size();
    cudaMalloc(&d_table, table_size * sizeof(uint64_t));
    cudaMalloc(&d_queries, keys.size() * sizeof(uint64_t));
    cudaMalloc(&d_results, keys.size() * sizeof(bool));

    cudaMemcpy(d_table, dict_table.data(), table_size * sizeof(uint64_t), cudaMemcpyHostToDevice);
    cudaMemcpy(d_queries, keys.data(), keys.size() * sizeof(uint64_t), cudaMemcpyHostToDevice);

    int blockSize = 256;
    int numBlocks = (keys.size() + blockSize - 1) / blockSize;
    lookup_gpu_multiple<<<numBlocks, blockSize>>>(d_table, table_size, d_queries, d_results, keys.size());

    std::vector<uint8_t> flat_results(keys.size());
    cudaMemcpy(flat_results.data(), d_results, keys.size() * sizeof(bool), cudaMemcpyDeviceToHost);
    
    std::vector<std::vector<bool>> results2D(n, std::vector<bool>(n, false));
    size_t idx = 0;
    for (size_t row = 0; row < n; ++row) {
        for (size_t col = row; col < n; ++col) {
            results2D[row][col] = flat_results[idx++] != 0;
        }
    }

    cudaFree(d_table);
    cudaFree(d_queries);
    cudaFree(d_results);

    return results2D;
}

std::string clean_string(const std::string &input) {
    std::string result;
    result.reserve(input.size()); 

    for (char c : input) {
        if (!std::ispunct(static_cast<unsigned char>(c))) { 
            result += std::tolower(static_cast<unsigned char>(c)); 
        }
    }

    return result;
}

void read_sentences_from_file(std::vector<std::string> &sentences) {
    FILE* f = fopen("long_sentences.txt", "r");
    char buffer[1024];

    if (!f) return;

    while (fgets(buffer, sizeof(buffer), f)) {
        buffer[strlen(buffer) - 1] = '\0';
        sentences.push_back(std::string(buffer));
    }

    fclose(f);
}

void read_mwes_from_file(std::vector<std::string> &mwe) {
    FILE* f = fopen("mwes.txt", "r");
    char buffer[1024];

    if (!f) return;

    while (fgets(buffer, sizeof(buffer), f)) {
        buffer[strlen(buffer) - 2] = '\0';
        mwe.push_back(std::string(buffer));
    }

    fclose(f);
}

int main() {
    std::vector<std::string> sample_sentences;
    read_sentences_from_file(sample_sentences);
    std::vector<std::string> mwes; 
    read_mwes_from_file(mwes);

    auto dict_table = build_dict_table(mwes);

    for (int i = 0; i < sample_sentences.size(); i++){
        auto sentence = sample_sentences[i];
        std::cout << "Processing - " << sentence << "\n";
        auto words = split_sentence(clean_string(sentence));
        
        auto lookup_table = create_lookup_table_gpu(words, dict_table);

        for (int x = 0; x < words.size(); x++){
            for (int y = 0; y < words.size(); y++){
                if (lookup_table[x][y] == 1){
                    std::cout << "Found MWE from index " << x << "-" << y << " - " << join_words(words, x, y) << "\n";
                }
            }
        }

        std::cout << "\n";
    }

    return 0;
}
```

```python
%%writefile lookup-gpuv2.cu

#include <array>
#include <cmath>
#include <cstdint>
#include <cstdio>
#include <iostream>
#include <memory>
#include <sstream>
#include <string>
#include <vector>
#include <stdlib.h>
#include <string.h>

static uint64_t* d_table_global = nullptr;
static size_t d_table_size_global = 0;

void init_dict_table_gpu(const std::vector<uint64_t>& dict_table)
{
    if (d_table_global != nullptr)
        return;

    d_table_size_global = dict_table.size();

    cudaMalloc(&d_table_global, d_table_size_global * sizeof(uint64_t));
    cudaMemcpy(d_table_global,
               dict_table.data(),
               d_table_size_global * sizeof(uint64_t),
               cudaMemcpyHostToDevice);
}

void free_dict_table_gpu()
{
    if (d_table_global != nullptr) {
        cudaFree(d_table_global);
        d_table_global = nullptr;
        d_table_size_global = 0;
    }
}

uint64_t fnv1a64(const std::string &s) {
    uint64_t hash = 1469598103934665603ULL; 
    for (unsigned char c : s) {
        hash ^= c;
        hash *= 1099511628211ULL; 
    }
    return hash ? hash : 1ULL; 
}

uint64_t next_power_of_two(uint64_t x) {
    return 1ULL << (64 - __builtin_clzll(x - 1));
}

std::vector<uint64_t> build_dict_table(const std::vector<std::string> &mwes) {
    size_t dict_count = mwes.size();
    size_t table_size = next_power_of_two((size_t)(dict_count / 0.6)); 
    std::vector<uint64_t> dict_table(table_size, 0ULL);

    for (const std::string &mwe : mwes) {
        uint64_t f = fnv1a64(mwe);
        size_t mask = table_size - 1;
        size_t idx = f & mask;
        while (dict_table[idx] != 0ULL) {
            idx = (idx + 1) & mask; 
        }
        dict_table[idx] = f;
    }

    std::cout << "Built hash table with " << table_size << " slots.\n";
    return dict_table;
}

__device__ bool lookup_gpu(const uint64_t *table, size_t table_size, uint64_t key) {
    size_t mask = table_size - 1;
    size_t idx = key & mask;
    for (int probe = 0; probe < 8; ++probe) { 
        uint64_t val = __ldg(&table[idx]);    
        if (val == 0ULL) return false;
        if (val == key) return true;
        idx = (idx + 1) & mask;
    }
    return false;
}

__global__ void lookup_gpu_multiple(const uint64_t *table, size_t table_size,
                                   const uint64_t *queries, bool *results, int n_queries) {
    int i = blockIdx.x * blockDim.x + threadIdx.x;
    int stride = blockDim.x * gridDim.x;
    
    for (; i < n_queries; i += stride)
        results[i] = lookup_gpu(table, table_size, queries[i]);
}

std::string join_words(const std::vector<std::string> &words, size_t start, size_t end) {
    std::ostringstream oss;
    for (size_t i = start; i <= end; ++i) {
        if (i > start) oss << " ";
        oss << words[i];
    }
    return oss.str();
}

std::vector<std::string> split_sentence(const std::string &sentence) {
    std::vector<std::string> words;
    std::istringstream iss(sentence);
    std::string word;
    
    while (iss >> word) {
        words.push_back(word);
    }
    
    return words;
}

std::vector<std::vector<bool>> create_lookup_table_gpu(
    const std::vector<std::string> &words)
{
    size_t n = words.size();
    std::vector<std::string> phrases;
    std::vector<uint64_t> keys;

    for (size_t row = 0; row < n; ++row) {
        for (size_t col = row; col < n; ++col) {
            std::string phrase = join_words(words, row, col);
            phrases.push_back(phrase);
            keys.push_back(fnv1a64(phrase));
        }
    }

    uint64_t *d_queries;
    bool *d_results;

    cudaMalloc(&d_queries, keys.size() * sizeof(uint64_t));
    cudaMalloc(&d_results, keys.size() * sizeof(bool));

    cudaMemcpy(d_queries, keys.data(), keys.size() * sizeof(uint64_t), cudaMemcpyHostToDevice);

    int blockSize = 256;
    int numBlocks = (keys.size() + blockSize - 1) / blockSize;

    lookup_gpu_multiple<<<numBlocks, blockSize>>>(
        d_table_global,
        d_table_size_global, 
        d_queries,
        d_results,
        keys.size()
    );

    std::vector<uint8_t> flat_results(keys.size());
    cudaMemcpy(flat_results.data(), d_results, keys.size() * sizeof(bool), cudaMemcpyDeviceToHost);

    std::vector<std::vector<bool>> results2D(n, std::vector<bool>(n, false));
    size_t idx = 0;
    for (size_t row = 0; row < n; ++row) {
        for (size_t col = row; col < n; ++col) {
            results2D[row][col] = flat_results[idx++] != 0;
        }
    }

    cudaFree(d_queries);
    cudaFree(d_results);

    return results2D;
}

std::string clean_string(const std::string &input) {
    std::string result;
    result.reserve(input.size()); 

    for (char c : input) {
        if (!std::ispunct(static_cast<unsigned char>(c))) { 
            result += std::tolower(static_cast<unsigned char>(c)); 
        }
    }

    return result;
}

void read_sentences_from_file(std::vector<std::string> &sentences) {
    FILE* f = fopen("long_sentences.txt", "r");
    char buffer[1024];

    if (!f) return;

    while (fgets(buffer, sizeof(buffer), f)) {
        buffer[strlen(buffer) - 1] = '\0';
        sentences.push_back(std::string(buffer));
    }

    fclose(f);
}

void read_mwes_from_file(std::vector<std::string> &mwe) {
    FILE* f = fopen("mwes.txt", "r");
    char buffer[1024];

    if (!f) return;

    while (fgets(buffer, sizeof(buffer), f)) {
        buffer[strlen(buffer) - 2] = '\0';
        mwe.push_back(std::string(buffer));
    }

    fclose(f);
}

int main() {
    std::vector<std::string> sample_sentences;
    read_sentences_from_file(sample_sentences);

    std::vector<std::string> mwes;
    read_mwes_from_file(mwes);

    auto dict_table = build_dict_table(mwes);
    init_dict_table_gpu(dict_table);

    for (int i = 0; i < sample_sentences.size(); i++) {
        auto sentence = sample_sentences[i];
        std::cout << "Processing - " << sentence << "\n";

        auto words = split_sentence(clean_string(sentence));
        auto lookup_table = create_lookup_table_gpu(words);

        for (int x = 0; x < words.size(); x++) {
            for (int y = 0; y < words.size(); y++) {
                if (lookup_table[x][y] == 1) {
                    std::cout << "Found MWE from index " 
                              << x << "-" << y << " - " 
                              << join_words(words, x, y) << "\n";
                }
            }
        }

        std::cout << "\n";
    }

    free_dict_table_gpu();
    return 0;
}
```

```python
%%writefile dp-gpuv2.cu
#include <array>
#include <climits>
#include <cmath>
#include <cstdint>
#include <cstdio>
#include <iostream>
#include <memory>
#include <sstream>
#include <string>
#include <vector>
#include <stdlib.h>
#include <string.h>

uint64_t fnv1a64(const std::string &s) {
    uint64_t hash = 1469598103934665603ULL;
    for (unsigned char c : s) {
        hash ^= c;
        hash *= 1099511628211ULL;
    }
    return hash ? hash : 1ULL;
}

uint64_t next_power_of_two(uint64_t x) {
    return 1ULL << (64 - __builtin_clzll(x - 1));
}

std::vector<uint64_t> build_dict_table(const std::vector<std::string> &mwes) {
    size_t dict_count = mwes.size();
    size_t table_size = next_power_of_two((size_t)(dict_count / 0.6)); 
    std::vector<uint64_t> dict_table(table_size, 0ULL);

    for (const std::string &mwe : mwes) {
        uint64_t f = fnv1a64(mwe);
        size_t mask = table_size - 1;
        size_t idx = f & mask;
        while (dict_table[idx] != 0ULL) {
            idx = (idx + 1) & mask; 
        }
        dict_table[idx] = f;
    }

    std::cout << "Built hash table with " << table_size << " slots.\n";
    return dict_table;
}

bool lookup_cpu(const std::vector<uint64_t> &table, uint64_t key) {
    size_t mask = table.size() - 1;
    size_t idx = key & mask;
    for (size_t probe = 0; probe < table.size(); ++probe) {
        uint64_t val = table[idx];
        if (val == 0ULL) return false;     
        if (val == key) return true;       
        idx = (idx + 1) & mask;            
    }
    return false; 
}

std::string join_words(const std::vector<std::string> &words, size_t start, size_t end) {
    std::ostringstream oss;
    for (size_t i = start; i <= end; ++i) {
        if (i > start) oss << " ";
        oss << words[i];
    }
    return oss.str();
}

std::vector<std::vector<bool>> create_lookup_table_cpu(const std::vector<std::string> &words,
                                    const std::vector<uint64_t> &dict_table) {
    size_t n = words.size();
    std::vector<std::vector<bool>> result(n, std::vector<bool>(n, false));

    for (size_t row = 0; row < n; ++row) {
        for (size_t col = row; col < n; ++col) { 
            std::string phrase = join_words(words, row, col);
            uint64_t key = fnv1a64(phrase);
            result[row][col] = lookup_cpu(dict_table, key);
        }
    }

    return result;
}


std::vector<std::string> split_sentence(const std::string &sentence) {
    std::vector<std::string> words;
    std::istringstream iss(sentence);
    std::string word;

    while (iss >> word) {
        words.push_back(word);
    }

    return words;
}

std::string clean_string(const std::string &input) {
    std::string result;
    result.reserve(input.size()); 

    for (char c : input) {
        if (!std::ispunct(static_cast<unsigned char>(c))) { 
            result += std::tolower(static_cast<unsigned char>(c)); 
        }
    }

    return result;
}

__global__
void compute_diagonal_mwe(const int* lookup_table, int* dp, int n) {
    int idx = threadIdx.x + blockIdx.x * blockDim.x;
    int stride = blockDim.x * gridDim.x;

    for (int x = idx; x < n; x += stride) {
        for (int i = 0; i < n; i++) {
            int y = x + i;
            if (y >= n) break;

            if (x == y || lookup_table[x * n + y]) {
                dp[x * n + y] = 1;
            } else {
                int min_count = INT32_MAX;
                for (int m = x; m < y; m++) {
                    int count = dp[x * n + m] + dp[(m + 1) * n + y];
                    if (count < min_count) min_count = count;
                }
                dp[x * n + y] = min_count;
            }
        }
        __syncthreads();
    }
}


int compute_minimum_mwes(const std::vector<std::string> &words, const std::vector<std::vector<bool>> &lookup_table) {
    const int n = words.size();
    
    int* d_lookup_table;
    int* d_dp;
    cudaMalloc(&d_lookup_table, n * n * sizeof(int));
    cudaMalloc(&d_dp, n * n * sizeof(int));
    
    std::vector<int> lookup_flat(n * n);
    for (int i = 0; i < n; i++) {
        for (int j = 0; j < n; j++) {
            lookup_flat[i * n + j] = lookup_table[i][j] ? 1 : 0;
        }
    }
    cudaMemcpy(d_lookup_table, lookup_flat.data(), n * n * sizeof(int), cudaMemcpyHostToDevice);
    
    std::vector<int> dp_flat(n * n, INT32_MAX);
    cudaMemcpy(d_dp, dp_flat.data(), n * n * sizeof(int), cudaMemcpyHostToDevice);

    compute_diagonal_mwe<<<1, n>>>(d_lookup_table, d_dp, n);
    cudaDeviceSynchronize();
    
    cudaMemcpy(dp_flat.data(), d_dp, n * n * sizeof(int), cudaMemcpyDeviceToHost);
    
    cudaFree(d_lookup_table);
    cudaFree(d_dp);

    return dp_flat[n - 1];
}

void read_sentences_from_file(std::vector<std::string> &sentences) {
    FILE* f = fopen("long_sentences.txt", "r");
    char buffer[1024];

    if (!f) return;

    while (fgets(buffer, sizeof(buffer), f)) {
        buffer[strlen(buffer) - 1] = '\0';
        sentences.push_back(std::string(buffer));
    }

    fclose(f);
}

void read_mwes_from_file(std::vector<std::string> &mwe) {
    FILE* f = fopen("mwes.txt", "r");
    char buffer[1024];

    if (!f) return;

    while (fgets(buffer, sizeof(buffer), f)) {
        buffer[strlen(buffer) - 2] = '\0';
        mwe.push_back(std::string(buffer));
    }

    fclose(f);
}

int main() {
    std::vector<std::string> sample_sentences;
    read_sentences_from_file(sample_sentences);
    std::vector<std::string> mwes; 
    read_mwes_from_file(mwes);

    auto dict_table = build_dict_table(mwes);

    for (int i = 0; i < sample_sentences.size(); i++){
        auto sentence = sample_sentences[i];
        std::cout << "Processing - " << sentence << "\n";
        auto words = split_sentence(clean_string(sentence));

        auto lookup_table = create_lookup_table_cpu(words, dict_table);

        std::cout << "Minimum MWEs: " << compute_minimum_mwes(words, lookup_table);

        std::cout << "\n";
    }

    return 0;
}


```
